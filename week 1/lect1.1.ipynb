- `GAN's` => Generative Adversarial Networks, it consists of two neural networks, the `generator and the discriminator`, that are trained simultaneously through adversarial training(meaning the generator tries to create realistic data to fool the discriminator, while the discriminator tries to distinguish between real and fake data)
    - adversarial means involving opposition or conflict, in the context of GANs, it refers to the competition between the generator and discriminator.
- `VAEs` => Variational Autoencoders, a type of generative model that `learns to encode input data into a latent space` and then decode it back to the `original space`, allowing for the generation of new data samples.
- `DDPMs` => `Denoising Diffusion Probabilistic Models, a class of generative models that `learn to generate data by reversing a gradual noising process`, effectively denoising data samples to create new, high-quality outputs.(score-based generative modeling)
- `Auto-Regressive Models` => A class of generative models that `generate data by predicting the next element in a sequence given the previous elements`, often used in natural language processing and time series forecasting.
