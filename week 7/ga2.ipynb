### Explanation:

In **Diffusion models** (specifically DDPM — Denoising Diffusion Probabilistic Models):

#### 🟠 **Forward Process**:

* This is the process of **adding noise** to data.
* It’s a **fixed Markov process**, usually with pre-defined noise schedule parameters (e.g., \$\beta\_t\$, \$\alpha\_t\$).
* These parameters are **not learnable**; they are **set** (e.g., linearly or cosine scheduled).

#### 🟢 **Reverse Process**:

* This is the process of **denoising** — learning to recover clean data from noise.
* This process is **parameterized by a neural network** (like a U-Net).
* These parameters **are learnable**, and training focuses on this.

---

Hence, only the **reverse process is learned**, not the forward process.

✅ **Correct: (C)**
