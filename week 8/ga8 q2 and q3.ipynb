# What is Forward Transition Probability?


In the context of **probabilistic models**, especially **sequential models** like **Hidden Markov Models (HMMs)**, **state-space models**, and **Variational Autoencoders (VAEs)**, the **forward transition probability** refers to the probability of moving from one state to the next in time.

---

### üîÅ **Definition:**

The **forward transition probability** is the conditional probability:

$$
q(x_t \mid x_{t-1})
$$

It represents the **probability distribution of the current state $x_t$** given the **previous state $x_{t-1}$**.

---

### üß† Intuition:

Think of a process evolving over time (like a particle moving, or a latent variable changing). The forward transition probability defines **how likely the next state is**, based on the current one.

---

### ‚úÖ Example (Gaussian Transition):

Suppose:

$$
q(x_t \mid x_{t-1}) = \mathcal{N}(\alpha_t x_{t-1}, 1 - \alpha_t^2)
$$

This means:

* The next state $x_t$ is **normally distributed**.
* Its **mean** depends linearly on the previous state ($\alpha_t x_{t-1}$).
* Its **variance** is $1 - \alpha_t^2$, which may shrink over time.

This kind of forward transition is often used in **diffusion models** or **VAEs with temporal dynamics**.

---

### üìå In Summary:

* **Forward transition probability**: $q(x_t \mid x_{t-1})$
* **Purpose**: Describes **how a state evolves** from time $t-1$ to $t$.
* **Used in**: HMMs, VAEs, Kalman filters, diffusion models, etc.
