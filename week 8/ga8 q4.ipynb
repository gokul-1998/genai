# What is DDPM?
**DDPM** stands for **Denoising Diffusion Probabilistic Models**. It's a **generative model** used to create data such as images, audio, etc., by learning to reverse a gradual noising process.

---

### 🔄 Key Idea:

DDPM works in **two phases**:

1. **Forward process (diffusion):** Gradually add Gaussian noise to a data sample over many steps, until it becomes pure noise.
2. **Reverse process (denoising):** Train a neural network to reverse the noising process, step by step, to reconstruct the original data from noise.

---

### 📦 Example (for images):

Imagine you take a clean image and slowly add Gaussian noise to it over 1000 steps, until it’s just static noise. DDPM trains a model to reverse this process—starting from noise and gradually denoising it back to a realistic image.

---

### 🧠 Why it works:

* The reverse process is modeled as a **Markov Chain**, with each step predicting the noise added in the forward process.
* The model is trained using **variational inference** and optimized using a **simplified loss function**: often, just an MSE loss between the true noise and predicted noise.

---

### 📚 Technical Details:

* Forward process:
  $q(x_t | x_{t-1}) = \mathcal{N}(\sqrt{\alpha_t}x_{t-1}, (1 - \alpha_t)I)$

* Reverse process is learned as:
  $p_\theta(x_{t-1} | x_t)$

* The model learns to predict the noise $\epsilon$ at each timestep using a neural network.

---

### 🎨 Applications:

* Image generation (e.g., DALL·E 2, Imagen)
* Super-resolution
* Inpainting
* Speech synthesis

---

### 🔥 Why it’s popular:

* Generates high-quality and diverse outputs.
* More stable than GANs.
* Used in **state-of-the-art models** like **Stable Diffusion**, **Imagen**, and **Midjourney**.
