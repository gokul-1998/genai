# What is Reverse Sampling

**Reverse sampling** refers to the process of generating data by starting from a noisy or latent representation and progressively **denoising** or **refining** it back to a more structured or meaningful form — often used in **generative models**, especially **Diffusion Models**.

---

### 🔄 In the context of **Diffusion Models** (like DDPM):

Reverse sampling is the **generation phase**.

---

### 🚀 Process Overview:

#### 1. **Forward Process (Diffusion)**

You add noise step-by-step to data $x_0$ over $T$ timesteps:

$$
x_0 \rightarrow x_1 \rightarrow x_2 \rightarrow \dots \rightarrow x_T \sim \mathcal{N}(0, I)
$$

By the final step, the data becomes pure noise.

#### 2. **Reverse Process (Sampling)**

Now you start from random noise $x_T \sim \mathcal{N}(0, I)$, and **reverse the noise** using a trained model:

$$
x_T \rightarrow x_{T-1} \rightarrow \dots \rightarrow x_0
$$

Each reverse step typically uses:

$$
x_{t-1} \sim q(x_{t-1} \mid x_t) \approx \mathcal{N}(\mu_t(x_t, \epsilon_\theta), \Sigma_t)
$$

Where $\epsilon_\theta$ is a model (like a U-Net) that predicts the noise component at each timestep.

---

### 🧠 Intuition:

* **Forward**: Add noise → data becomes meaningless
* **Reverse**: Denoise step-by-step → recover realistic data (like an image)

---

### 📌 Applications:

* Generating realistic images (e.g., DALL·E 2, Stable Diffusion)
* Audio generation
* Text-to-image models
